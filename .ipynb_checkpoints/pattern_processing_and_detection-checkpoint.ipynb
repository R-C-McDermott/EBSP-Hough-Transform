```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import PIL
from PIL import Image

import skimage.io
from skimage.transform import hough_line, hough_line_peaks
from skimage.feature import canny
from skimage import data
from skimage.color import rgb2gray, rgba2rgb
from skimage.filters import threshold_otsu
from skimage.filters import threshold_local

from scipy.ndimage.filters import correlate
from scipy import stats
from scipy.signal import argrelextrema, find_peaks

import matplotlib.pyplot as plt
from matplotlib import cm

from aloe.plots import plot_image
from aloe.image.downsample import downsample
from aloe.image.kikufilter import make_signal_bg_fft, remove_bg_fft
```

```python
# Loading ten experimental patterns from same grain and static background captured in low magnification mode

raw_EBSD0 = Image.open("1_70.tiff")
raw_EBSD1 = Image.open("1_71.tiff")
raw_EBSD2 = Image.open("1_72.tiff")
raw_EBSD3 = Image.open("1_73.tiff")
raw_EBSD4 = Image.open("1_74.tiff")
raw_EBSD5 = Image.open("1_75.tiff")
raw_EBSD6 = Image.open("1_76.tiff")
raw_EBSD7 = Image.open("1_77.tiff")
raw_EBSD8 = Image.open("1_78.tiff")
raw_EBSD9 = Image.open("1_79.tiff")
raw_EBSD10 = Image.open("1_80.tiff")

static_BG = Image.open("StaticBackground.tiff")
```

```python
# Convert static BG to 256-bit float
BG_255 = 255*(np.array(static_BG)/np.max(static_BG))
maximum_intensity = np.max(BG_255)

# Pattern centre recorded from software
position_x, position_y = 0.49493*BG_255.shape[1], 0.17443*BG_255.shape[0]
print(f"X: {position_x} Y: {position_y}")

plt.imshow(BG_255, cmap="gray")
plt.plot(position_x, position_y, "b+")
plt.show()

plt.imshow(raw_EBSD0, cmap="gray")
plt.plot(position_x, position_y, "b+")
```

```python
plot_image(raw_EBSD0, title="Raw EBSP")
```

```python
## Pattern averaging ##

pattern_list = [raw_EBSD0, raw_EBSD1, raw_EBSD2, raw_EBSD3, raw_EBSD4, raw_EBSD5,
                raw_EBSD6, raw_EBSD7, raw_EBSD8, raw_EBSD9, raw_EBSD10]

pattern_list_arr = [255*(np.array(pattern)/np.max(pattern)) for pattern in pattern_list]

pattern_avg = sum(pattern_list_arr)/11

# Background array with slight intensity correction
BG_arr = 205*(np.array(static_BG)/np.max(static_BG))+20


# Print image shapes and plot images
print(BG_arr.shape)
print(pattern_avg.shape)
plot_image(pattern_avg, title="Ten-pattern average scaled to 8-bit")
plot_image(BG_arr, title="Static background scaled to 8-bit")

# Intensity histograms
n, bins, patches = plt.hist(np.ravel(pattern_avg), 256, range=[0,256])
plt.title('histogram of raw pattern')
plt.show()

n, bins, patches = plt.hist(np.ravel(BG_arr), 256, range=[0,256])
plt.title('histogram of static background')
plt.show()
```

```python
# Background subtraction - testing different approaches for background correction

BG_subtraction = pattern_avg - BG_arr
BG_subtraction = BG_subtraction+(-1)*np.min(BG_subtraction)
BG_subtraction = 255*(BG_subtraction/np.max(BG_subtraction))

BG_subtraction_with_div = (pattern_avg - BG_arr)/BG_arr
BG_subtraction_with_div = BG_subtraction_with_div+(-1)*np.min(BG_subtraction_with_div)
BG_subtraction_with_div = 255*(BG_subtraction_with_div/np.max(BG_subtraction_with_div))

BG_subtraction_with_div_only = pattern_avg/BG_arr
BG_subtraction_with_div_only = BG_subtraction_with_div+(-1)*np.min(BG_subtraction_with_div)
BG_subtraction_with_div_only = 255*(BG_subtraction_with_div/np.max(BG_subtraction_with_div))


plot_image(BG_subtraction, title="BG subtracted")
plot_image(BG_subtraction_with_div, title="BG subtracted with divide")
plot_image(BG_subtraction_with_div_only, title="BG divide")


n, bins, patches = plt.hist(np.ravel(BG_subtraction), 256, range=[0,256])
plt.title('histogram of BG subtracted')
plt.show()

n, bins, patches = plt.hist(np.ravel(BG_subtraction_with_div), 256, range=[0,256])
plt.title('histogram of BG subtracted with divide')
plt.show()

n, bins, patches = plt.hist(np.ravel(BG_subtraction_with_div_only), 256, range=[0,256])
plt.title('histogram of pattern with static BG correction')
plt.show()
```

```python
# FFT - Capture dynamic background

raw_pattern_signal, raw_pattern_bg = make_signal_bg_fft(BG_subtraction_with_div, sigma=50, bgscale=1.0)
plot_image(raw_pattern_signal, [-60, 60])
plot_image(raw_pattern_bg)
plot_image(BG_subtraction_with_div)
```

```python
# Substracting dynamic background from static background corrected pattern


new_bg_sub = (BG_subtraction_with_div-raw_pattern_bg)
new_bg_sub = new_bg_sub+(-1)*np.min(new_bg_sub)
new_bg_sub = 255*(new_bg_sub/np.max(new_bg_sub))

plot_image(new_bg_sub)

n, bins, patches = plt.hist(np.ravel(raw_pattern_bg), 256, range=[0,256])
plt.title('histogram of dynamic background')
plt.show()

n, bins, patches = plt.hist(np.ravel(new_bg_sub), 256, range=[0,256])
plt.title('histogram of intensity after dynamic background subtraction')
plt.show()

```

```python
# Corrected image including PC

plt.imshow(new_bg_sub, cmap="gray")
plt.plot(position_x, position_y, "r+", markersize=15, markeredgewidth=2)
plt.show()
print(new_bg_sub.shape)
```

```python
# Contrast stretching

p_low, p_high = np.percentile(new_bg_sub, (3, 97)) # Stretch to 3rd and 97th percentiles

img_contrast_stretch = skimage.exposure.rescale_intensity(new_bg_sub, in_range=(p_low, p_high))
img_contrast_stretch *= 255
plot_image(img_contrast_stretch, [0,255])

n, bins, patches = plt.hist(np.ravel(img_contrast_stretch), 256, range=[1,254])
plt.title('histogram of contrast stretch')
plt.show()
```

```python
# Image binning

binning=4
img_binned=downsample(img_contrast_stretch, binning)
plot_image(img_binned)
```

```python
# Threshold image - Binarize

factor = 1.15
global_thresh = factor*threshold_otsu(img_binned)
binary = img_binned > global_thresh

fig, axes = plt.subplots(nrows=2, figsize=(7, 8))
ax0, ax1 = axes
plt.gray()

ax0.imshow(img_binned)
ax0.set_title('Binned image')

ax1.imshow(binary)
ax1.set_title('Global thresholding')

for ax in axes:
    ax.axis('off')
    
plt.show()
```

```python
# Classic straight-line Hough transform
# Set a precision of 0.5 degree.

# Hough space too noisy for band detection!

tested_angles = np.linspace(-np.pi / 2, np.pi / 2, 360, endpoint=False)
h, theta, d = hough_line(binary, theta=tested_angles)

hpeaks=hough_line_peaks(h, theta, d,
        min_distance=10, min_angle=15,
        threshold=0.5*np.max(h), num_peaks=10)

# Generating figure 1
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(binary, cmap=cm.gray)
ax[0].set_title('Input image')
# ax[0].set_axis_off()

angle_step = 0.5 * np.diff(theta).mean()
d_step = 0.5 * np.diff(d).mean()
bounds = [np.rad2deg(theta[0] - angle_step),
          np.rad2deg(theta[-1] + angle_step),
          d[-1] + d_step, d[0] - d_step]
ax[1].imshow(np.log(1 + h), extent=bounds, cmap=cm.gray, aspect=1 / 1.5)
ax[1].set_title('Hough transform')
ax[1].set_xlabel('Angles (degrees)')
ax[1].set_ylabel('Distance (pixels)')
ax[1].axis('image')

ax[2].imshow(img_contrast_stretch, cmap=cm.gray)
ax[2].set_ylim((img_contrast_stretch.shape[0]-1, 0))
ax[2].set_xlim((img_contrast_stretch.shape[1], 0))

ax[2].set_axis_off()
ax[2].set_title('Detected lines')

for _, angle, dist in zip(*hpeaks):
    (x0, y0) = binning*dist * np.array([np.cos(angle), np.sin(angle)])
    ax[2].axline((x0, y0), slope=np.tan(angle + np.pi/2), color='r')

plt.tight_layout()
plt.show()
```

```python
# Classic straight-line Hough transform from binary data - With butterfly convolution

image=np.flipud(binary)
background=np.ones_like(image)

ang=np.deg2rad(np.linspace(-90,90,360))
h, theta, d = hough_line(image, theta=ang)
h_masked = np.ma.masked_equal(h ,0)
h_masked=h_masked/np.max(h_masked)

# butterfly mask convolution
# 9x9 mask Krieger-Lassen
k = np.array([[-10, -15, -22, -22, -22, -22, -22, -15, -10],
              [  1,  -6, -13, -22, -22, -22, -13,  -6,   1],
              [  3,   6,   4,  -3, -22,  -3,   4,   6,   3],
              [  3,  11,  19,  28,  42,  28,  19,  11,   3],
              [  3,  11,  27,  42,  42,  42,  27,  11,   3],
              [  3,  11,  19,  28,  42,  28,  19,  11,   3],
              [  3,   6,   4,  -3, -22,  -3,   4,   6,   3],
              [  1,  -6, -13, -22, -22, -22, -13,  -6,   1],
              [-10, -15, -22, -22, -22, -22, -22, -15, -10] ])

h = correlate(h_masked, k, mode='nearest') #.astype(np.int64)
a=(h-np.min(h))/(np.max(h)-np.min(h))

hpeaks=hough_line_peaks(a, theta, d,
        min_distance=10, min_angle=15,
        threshold=0.5*np.max(a), num_peaks=8)
```

```python
# Band detection with butterfly convolution applied to Hough space.... much better results!


fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(6, 13))

ax0.imshow(image)
ax0.set_title('Processed Input Data')
ax0.set_axis_off()

#ax1.imshow(np.log(1 + h), extent=[np.rad2deg(theta[-1]),
#            np.rad2deg(theta[0]),
#           d[-1], d[0]], cmap=cm.gray, aspect=1)
ax1.imshow(a,cmap=cm.gray, extent=[np.rad2deg(ang[-1]), np.rad2deg(ang[0]), d[-1], d[0]])
ax1.set_title('Hough transform')
ax1.set_xlabel('Angles (degrees)')
ax1.set_ylabel('Distance (pixels)')
ax1.axis('image')

ax2.imshow(np.flipud(img_contrast_stretch), cmap=cm.gray)

row1, col1 = img_contrast_stretch.shape
linecount=1
houghkl=[]
houghlines=[]
for _, angle, dist in zip(*hpeaks):
    y0 = (dist*binning - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist*binning - col1 * np.cos(angle)) / np.sin(angle)
    ax2.plot((0, col1), (y0, y1), '-r')
    houghlines.append([linecount, dist, np.degrees(angle),
                       0, y0, col1,y1 ])
    #print(linecount, dist, np.rad2deg(angle),gpc)
    linecount=linecount+1


ax2.axis((col1, 0, 0, row1))

# ax2.axis((0, col1, row1, 0))
ax2.set_title('detected lines')
ax2.set_axis_off()
plt.tight_layout()
plt.show()
```
